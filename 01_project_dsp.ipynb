```python
# import libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

%matplotlib inline
```

```python
# importing data

sdf = pd.read_csv('students_performance.csv')
```

#### Create a new column 'total score' by summing up the math, reading, and writing scores.

```python
sdf['total score'] = sdf['math score']+ sdf['reading score'] + sdf['writing score']
```

```python
sdf
```

#### Create a new column 'pass/fail' to indicate whether a student passed or failed based on a score threshold of 60.

```python
sdf['pass/fail'] = np.where(sdf['total score'] >= 180, 'Pass', 'Fail')
```

```python
sdf
```

#### compare the average math scores across different genders.

```python
# creat a string for mean math based on gender 

avrg_math = sdf.groupby('gender')['math score'].mean()
```

```python
# creating bar plot

plt.figure(figsize=(8, 6))
plt.bar(avrg_math.index, avrg_math, color='steelblue')

plt.title('Average Math Scores by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Math Score', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')

plt.show()
```

#### compare the distribution of reading scores across different parental levels of education.

```python
# we can use 'writing score', 'math score' as well


plt.figure(figsize=(10, 6))

sns.boxplot(x='parental level of education', y='reading score', data=sdf, palette='Set3')

plt.title('Distribution of Reading Scores by Parental Level of Education', fontsize=16)
plt.xlabel('Parental Level of Education', fontsize=12)
plt.ylabel('Reading Score', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')
```

#### Compare 'total score' across different parental levels of education, to check how the parent level of education affects students overall performance  

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x='parental level of education', y='total score', data=sdf, palette='Set3')

plt.title('Overall score Vs Parental Level of Education', fontsize=16)
plt.xlabel('Parental Level of Education', fontsize=12)
plt.ylabel('total Score', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')
```

#### Calculating the correlation between math, reading, writing, and total scores.

```python
corl = sdf[['math score', 'reading score', 'writing score', 'total score']].corr()
```

```python
corl
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corl, annot=True, cmap='RdBu_r', center=0, linewidths=0.5, cbar=False, square=True)

plt.title('Correlation Matrix', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Create a separate colorbar
cax = plt.imshow(np.array([[0, 1]]), cmap='RdBu_r')
plt.colorbar(cax, orientation='vertical', pad=0.05)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Define colors for boys and girls
colors = {'male': 'lightblue', 'female': 'mistyrose'}

# Plot the bubble plot
sns.scatterplot(x='race/ethnicity', y='total score', hue='gender', size='total score',
                palette=colors, data=sdf)

plt.title('Total Score by Race/Ethnicity and Gender', fontsize=16)
plt.xlabel('Race/Ethnicity', fontsize=12)
plt.ylabel('Total Score', fontsize=12)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')



# Move the legend outside the plot
plt.legend(fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')


plt.show()
```
